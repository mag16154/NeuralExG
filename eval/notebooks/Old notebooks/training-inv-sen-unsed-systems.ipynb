```python
import matplotlib
%matplotlib inline

import sys
sys.path.append('../../core')
sys.path.append('../../configuration-setup/')

from learningModule import DataConfiguration, CreateTrainNN

epochs = 20
dynamics = 'InvPendulumD'
dataObject = DataConfiguration(dynamics=dynamics, dimensions=4, gradient_run=True)
dataObject.setSteps(200)
dataObject.setSamples(70)
dataObject.setLowerBound([-0.5, 0.0, -0.2, 0.0])
dataObject.setUpperBound([0.5, 0.0, 0.2, 0.0])
dataObject.generateTrajectories(scaling=0.001)
dataObject.showTrajectories(xindex=0, yindex=2)
dataObject.createData(jumps=[1,2,4,8,10,12,16])

nnObject1 = CreateTrainNN(dynamics=dynamics, dnn_rbf='dnn')
nnObject1.createInputOutput(data_object = dataObject, inp_vars=['x','xp','vp','t'], out_vars=['v'])
nnObject1.setEpochs(epochs)
nnObject1.trainTestNN(optim='SGD', loss_fn='mae', layers=2, neurons=128, act_fn='Tanh')
```
