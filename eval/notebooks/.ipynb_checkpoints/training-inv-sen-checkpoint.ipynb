```python
# Trained with the supplementary data just for testing
import matplotlib
%matplotlib inline

import sys
sys.path.append('../../core')
sys.path.append('../../configuration-setup/')

from learningModule import DataConfiguration, CreateTrainNN

epochs=30
dynamics = 'Brussellator'
dataObject = DataConfiguration(dynamics=dynamics, gradient_run=True)
dataObject.setSteps(250)
dataObject.setSamples(50)
dataObject.setLowerBound([0.0, 0.0])
dataObject.setUpperBound([2.0, 2.0])
dataObject.generateTrajectories(scaling=0.01)
dataObject.showTrajectories()
dataObject.createData(jumps=[1,2,4,8,10,12,16])

nnObject1 = CreateTrainNN(dynamics=dynamics, dnn_rbf='RBF')
nnObject1.createInputOutput(data_object = dataObject, inp_vars=['x','xp','vp','t'], out_vars=['v'])
nnObject1.setEpochs(epochs)
nnObject1.trainTestNN(optim='SGD', loss_fn='mae', layers=1, neurons=256)
```
