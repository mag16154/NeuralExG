```python
import matplotlib
%matplotlib inline
import numpy as np

import sys
sys.path.append('../../core')
sys.path.append('../../configuration-setup/')
sys.path.append('../../core/rrtAlgorithms/src/')

from evaluation import Evaluation

evalObject = Evaluation(dynamics='Brussellator')
evalObject.setIterCount(500)
dataObject = evalObject.setDataObject()
# For 35
# dataObject.setLowerBound([0.5, 0.2])
# dataObject.setUpperBound([0.8, 0.5])
# lowerBound = [0.53, 0.6]
# upperBound = [0.55, 0.7]
# evalObject.setUnsafeSet(lowerBound, upperBound)

dataObject.setLowerBound([0.2, 1.1])
dataObject.setUpperBound([1.5, 1.4])
lowerBound = [0.58, 1.70]
upperBound = [0.61, 1.72]
evalObject.setUnsafeSet(lowerBound, upperBound)
dest_traj = dataObject.generateTrajectories(samples=1)[0]
time_step = 110

# print(dest_traj[0], dest_traj[time_step])
# dest = dest_traj[time_step]
# dests = [dest]
# print(dests)
max_iter = 500
f_iterations = np.zeros(max_iter)
min_rel_dists = np.zeros(max_iter)
idx = 0
while idx < max_iter:
    ref_traj = dataObject.generateTrajectories(samples=1)[0]
    print(ref_traj[0], ref_traj[time_step])
    dest = evalObject.generateRandomUnsafeStates(1)
    dests = dest
    print(dests)
    f_iter, rel_dist = evalObject.reachDestInvPaths(dests=dests, d_time_steps=[time_step], threshold=0.01, 
                            correction_steps=[10],
                            scaling_factors=[0.1], i_state=[ref_traj[0]])
    if f_iter > 1 and rel_dist < 1.0:
        f_iterations[idx], min_rel_dists[idx] = f_iter, rel_dist
        idx = idx + 1
    else:
        print("** either f_iter is 1 or rel_dist is 1.0 **")
# print(f_iterations, min_rel_dists)
print(np.sum(f_iterations)/max_iter)
print(np.sum(min_rel_dists)/max_iter)
```
