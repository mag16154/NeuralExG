```python
import matplotlib
%matplotlib inline
import numpy as np

import sys
sys.path.append('../../core')
sys.path.append('../../configuration-setup/')
sys.path.append('../../core/rrtAlgorithms/src/')

from evaluation import Evaluation

evalObject = Evaluation(dynamics='OtherBenchC10')
evalObject.setIterCount(21)
dataObject = evalObject.setDataObject()

lowerBound = [7.95, -3.4, 2.5, 2.3]
evalObject.setUnsafeSet(lowerBound, upperBound)
dest_traj = dataObject.generateTrajectories(samples=1)[0]
time_step = 91

dest = [dest_traj[time_step]]

print(dest_traj[0], dest)

max_iter = 500
f_iterations = np.zeros(max_iter)
min_dists = np.zeros(max_iter)
idx = 0
while idx < max_iter:
    print("Idx *** " + str(idx))
    ref_traj = dataObject.generateTrajectories(samples=1)[0]
    print(ref_traj[0], ref_traj[time_step])
#     dest = evalObject.generateRandomUnsafeStates(1)
    dests = dest
    print(dests)
    evalObject.reachDestInvPaths(dests=dests, d_time_steps=[time_step], threshold=0.005, 
                            correction_steps=[20],
                            scaling_factors=[0.01], i_state=[ref_traj[0]], neurons = 512, adaptation_run=True)
    f_iter = evalObject.getFIterations()
    min_dist = evalObject.getFDistance()
    rel_dist = evalObject.getFRelDistance()
    if f_iter > 1 and f_iter < 100 and rel_dist < 1.0:
        f_iterations[idx], min_dists[idx] = f_iter, min_dist
        idx = idx + 1
    else:
        print("** either f_iter is 1 or rel_dist is 1.0 **")
# print(f_iterations, min_rel_dists)
print(np.sum(f_iterations)/max_iter)
print(np.sum(min_dists)/max_iter)
print(np.max(f_iterations))
print(np.var(f_iterations))
```
